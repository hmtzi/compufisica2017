```julia
x=collect(0.0:1.0:10.0);
y=zeros(x);
m=2.0;
b=1.0;
for i in eachindex(x)
    y[i]=m*x[i]+b
end
```

```julia
T1=["x"]
T2=["y=f(x)"]
T=hcat(T1,T2)
```

```julia
Tabla0=hcat(x,y)
```

```julia
Tabla=vcat(T,Tabla0)
```

```julia
writedlm("ValoresY.txt",Tabla,'|')
```

```julia
;cat ValoresY.txt
```

```julia
using Interact
```

```julia
@manipulate for m=0.0:1.0:4.0, b=0.0:1.0:4.0
    y=m.*x.+b
    plot(x,y,w=2,xlim=(minimum(x),maximum(x)),ylim=(minimum(y),maximum(y)),label=["y=f(x)"],lcolor=:darkblue)
    title!("Ecuación de una recta")
    xaxis!("x")
    yaxis!("y")
end
```
